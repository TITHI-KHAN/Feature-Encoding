```python
#1st Way

for col in df6.columns:
  if is_numeric_dtype(df6[col]):
    continue
  else:
    one=pd.get_dummies(df6[col],drop_first=True, prefix='One_Hot_')
    df6.drop(col, axis=1, inplace=True) #drops a column from the DataFrame and inplace=True means the work will be automatically updated in df6.  I won't have to write like this: df6=.......
    df6=pd.concat([df6, one], axis=1)
```

```python
df6.head()
```

```python
#2nd Way

from pandas.core.dtypes.common import is_numeric_dtype
for col in df7.columns:
  if not is_numeric_dtype(df7[col]):
    one=pd.get_dummies(df7[col],drop_first=True, prefix='One_Hot_'+ col)
    print(one)
```

```python
one.head()
```

```python
#3rd Way

from pandas.core.dtypes.common import is_numeric_dtype

one=pd.DataFrame() # Create an empty DataFrame

for col in df8.columns:
  if not is_numeric_dtype(df8[col]):
    temp=pd.get_dummies(df8[col],drop_first=True, prefix='One_Hot_'+ col) #the dummy variables for each non-numeric column are stored in the temporary DataFrame temp.
    one=pd.concat([one,temp], axis=1) #This ensures that all non-numeric columns are included in the resulting one DataFrame.

print(one.head())
```

Ordinal Encoding using Loop

```python
df11.head()
```

```python
#1st Way

from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype 

oe=OrdinalEncoder()

for col in df11.columns:
  if not is_numeric_dtype(df11[col]):
    df11[col]=oe.fit_transform(df11[[col]])

#The double brackets [[]] are used to ensure that the input to the fit_transform() method of OrdinalEncoder is a 2D array or DataFrame.
#df11[col] is a Pandas Series representing a single column of the DataFrame df11. To convert it into a 2D array, you need to wrap it with double brackets, like df11[[col]], which creates a DataFrame with a single column.
```

```python
df11.head()
```

```python
#2nd Way

from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype 

oe=OrdinalEncoder()

for col in df11.columns:
  if is_numeric_dtype(df11[col]):
    continue
  df11[col]=oe.fit_transform(df11[[col]])
```

```python
#3rd Way

from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype 

oe=OrdinalEncoder()

for col in df11.columns:
  if is_numeric_dtype(df11[col]):
    continue
  else:
    df11[col]=oe.fit_transform(df11[[col]])
```
